Tentative de récupération des données météo via OpenWeatherMap

```python
import json
import requests
```

```python
base_url = "https://api.open-meteo.com/v1/meteofrance?"
latitude=str(52.52)
longitude=str(13.41)

complete_url = base_url + "latitude=" + latitude + "&longitude=" + longitude + "&hourly=windspeed_10m"
print(complete_url)
reponse = requests.get(complete_url)
print(reponse.content)
```
