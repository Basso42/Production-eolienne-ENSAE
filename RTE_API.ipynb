Avant de pouvoir interagir avec l'API de RTE, on doit aller récupérer ses jetons d'authentification (donnés gratuitement après création d'un compte sur la plateforme de RTE).

#### On récupère un token d'authentification OAuth2

```python
client_id = "ac96732a-51ba-4487-91ee-755f22119c52"
client_secret = "12640de5-1bdf-411e-bc9d-905776c3557e"
url = "https://digital.iservices.rte-france.com/token/oauth/"
```

```python
def get_access_token(url, client_id, client_secret):
    response = requests.post(
        url,
        data={"grant_type": "client_credentials"},
        auth=(client_id, client_secret),
    )
    return response.json()["access_token"]


myToken = get_access_token(url, client_id, client_secret)
```

#### On appelle l'API de production d'électricité de RTE avec le jeton OAuth2
